```python
from sqlalchemy import create_engine

conn_url = 'postgresql://postgres:123@localhost/lovoo'
engine = create_engine(conn_url)
connection = engine.connect()
connection.autocommit = True
```

```python
stmt_view_drop = """
DROP TABLE User_charm;
DROP TABLE User_Genderlooking;
DROP TABLE User_purpose;
DROP TABLE User_language;
DROP TABLE User_socialmedia;
DROP TABLE Notification_record;
DROP TABLE Video_record;
DROP TABLE User_match;
DROP TABLE Current_location;
DROP TABLE Photo;

DROP TABLE User_profile;
DROP TABLE Genderlooking;
DROP TABLE Purpose;
DROP TABLE Language;
DROP TABLE Socialmedia;
DROP TABLE Message;
"""
connection.execute(stmt_view_drop)
```

```python
stmt = """
CREATE TABLE User_profile (
	User_id 				varchar(100),
	name 					varchar(50),
	age 					smallint,
	gender 					varchar(10),
	whazzup 				varchar(255),
	Country 				varchar(50),
	City 					varchar(50),
	isvip 					boolean,
	isnew 					boolean,
  	PRIMARY KEY (User_id)
);


CREATE TABLE Current_location (
	Current_location_id		varchar(50),
	User_id 				varchar(100),
	Current_location 		varchar(50),
	distance 				numeric(6,2),
	location_timestamp		timestamp,
  	PRIMARY KEY (Current_location_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id)
);

CREATE TABLE Photo (
	Photo_id 				varchar(50),
	User_id 				varchar(100),
	Photo_url 				varchar(255),
  	PRIMARY KEY (Photo_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id)
);

CREATE TABLE Video_record (
	Video_record_id 		varchar(50),
	User_id 				varchar(100),
	playing_timestamp		timestamp,
	video_format 			varchar(20),
	video_url 				varchar(255),
  	PRIMARY KEY (Video_record_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id)
);

CREATE TABLE Notification_record (
	Notification_record_id 	varchar(50),
	User_id 				varchar(100),
	Notification_timestamp	timestamp,
	Notification_content 			varchar(50),
  	PRIMARY KEY (Notification_record_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id)
);

CREATE TABLE Genderlooking (
	Genderlooking_id 		varchar(50),
	Genderlooking			varchar(10),
	PRIMARY KEY (Genderlooking_id)
);

CREATE TABLE Purpose (
	Purpose_id 				varchar(50),
	Purpose 				varchar(50),
  	PRIMARY KEY (Purpose_id)
);

CREATE TABLE Language (
	Language_id 			varchar(50),
	Language 				varchar(5),
	PRIMARY KEY (Language_id)
);

CREATE TABLE Socialmedia (
	Socialmedia_id 			varchar(50),
	Socialmedia 			varchar(10),
	PRIMARY KEY (Socialmedia_id)
);

CREATE TABLE User_charm (
	User_id 				varchar(100),
	profileVisit_number 	integer,
	fans_number 			integer,
	kisses_number 			integer,
	isFlirtstar 			boolean,
	isHighlighted 			boolean,
	isInfluencer 			boolean,
  	PRIMARY KEY (User_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id)
);


CREATE TABLE User_Genderlooking (
	User_id 				varchar(100),
	Genderlooking_id 		varchar(50),
  	PRIMARY KEY (User_id, Genderlooking_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id),
	FOREIGN KEY (Genderlooking_id) REFERENCES Genderlooking (Genderlooking_id)
);

CREATE TABLE User_purpose (
	User_id 				varchar(100),
	Purpose_id	 			varchar(50),
  	PRIMARY KEY (User_id, Purpose_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id),
	FOREIGN KEY (Purpose_id) REFERENCES Purpose (Purpose_id)
);

CREATE TABLE User_language (
	User_id					varchar(100),
    Language_id				varchar(50),
  	PRIMARY KEY (User_id,Language_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id),
	FOREIGN KEY (Language_id) REFERENCES Language (Language_id)
);

CREATE TABLE User_socialmedia (
	User_id					varchar(100),
    Socialmedia_id			varchar(50),
  	PRIMARY KEY (User_id,Socialmedia_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id),
	FOREIGN KEY (Socialmedia_id) REFERENCES Socialmedia (Socialmedia_id)
);

CREATE TABLE Message (
	Message_id 				varchar(50),
	Message_content 		varchar(255),
	Message_timestamp 		timestamp,
	PRIMARY KEY (Message_id)
);


CREATE TABLE User_match (
	Match_id 				varchar(50),
	User_id 				varchar(100),
	Matched_user_id 		varchar(100),
	Message_id 				varchar(50),
	PRIMARY KEY (Match_id, Message_id),
	FOREIGN KEY (User_id) REFERENCES User_profile (User_id),
	FOREIGN KEY (Matched_user_id) REFERENCES User_profile (User_id),
	FOREIGN KEY (Message_id) REFERENCES Message (Message_id)
);



"""
connection.execute(stmt)
```
